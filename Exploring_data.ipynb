```python
import pandas as pd
import statistics as stat
pd.set_option('display.max_rows', 100)
```

```python
infile = './../data/dataset_mood_smartphone.csv'
```

```python
df = pd.read_csv(infile, index_col=0)
df.sort_values(['time'])
```

```python
df.loc[df['variable'] == 'mood'].loc[df['id']=='AS14.01'].sort_values(['time'])[:100]
```

```python
for variable in set(df['variable']):
    values = df.loc[df['variable'] == variable]['value']
    total =  len(values)
    print(f"""{variable}
    total: {total}
    mean: {values.mean()}
    median: {values.median()}
    median counts: {len(df.loc[df['variable'] == variable].loc[df['value'] == values.median()])}
    median percentage: {(len(df.loc[df['variable'] == variable].loc[df['value'] == values.median()]))*100/total}
    lowest: {values.min()}
    lowest counts: {len(df.loc[df['variable'] == variable].loc[df['value'] == values.min()])}
    lowest percentage: {(len(df.loc[df['variable'] == variable].loc[df['value'] == values.min()]))*100/total}
    highest: {values.max()}
    highest counts: {len(df.loc[df['variable'] == variable].loc[df['value'] == values.max()])}
    highest percentage: {(len(df.loc[df['variable'] == variable].loc[df['value'] == values.max()]))*100/total}
    """)
```
